```python
centers, conditions = get_centers( n_cluster=5, method='variance', filetype='masked', max_k=10)
```

```python
node_number=[6,5,8,10,4,5,7]
node_networks = ['DAN','VAN','SMN','VIS','AUD','LAN','DMN']
position = [sum(node_number[:i+1]) for i in range(len(node_number))]
position_label = [-0.5+position[i]-node_number[i]/2. for i in range(len(node_number))]
```

```python
order_movie = np.array([2,1,3,5,4]) - 1
order_scramble = np.array([3,5,2,1,4]) - 1
centers = np.array(centers)
centers[0] = centers[0][order_movie]
centers[1] = centers[1][order_scramble]
centers[2] = centers[2][range(5)]
```

```python
fig = pl.figure(figsize=(16,13))
for i, center in enumerate(centers):
    for j, matrix in enumerate(center):
        ax = fig.add_subplot(3, 5, i*5+j+1)
        matrix = copy_matrix(array_to_matrix(matrix), diagonal_filler=0)
        total_nodes = matrix.shape[0]
        ax.imshow(matrix, interpolation='nearest', vmin=0, vmax=1)
        for name, n_nodes in zip(node_networks, position):
            ax.vlines(n_nodes-0.5, -0.5, total_nodes-0.5)
            ax.hlines(n_nodes-0.5, -0.5, total_nodes-0.5)
        ax.set_title(conditions[i]+' state '+str(j+1))
        ax.set_xticks(position_label)
        ax.set_xticklabels(node_networks, rotation=45)
        ax.set_yticks(position_label)
        ax.set_yticklabels(node_networks)
```

State matrix for each condition.

```python
centroid_ts = TimeSeries(np.vstack(centers[0]), sampling_interval=1)
S = SimilarityAnalyzer(centroid_ts, measure=correlation)
```

```python
similarity = 1 - S.measure # Transformation from correlation distance to correlation coefficient
```

```python
pl.figure(figsize=(10,8))
pl.imshow(copy_matrix(similarity, diagonal_filler=1), cmap=pl.cm.RdBu_r)
pl.title("Correlation of Movie states")
pl.xticks(np.arange(5),["Movie state %s" % str(i+1) for i in range(5)])
pl.yticks(np.arange(5),["Movie state %s" % str(i+1) for i in range(5)])
_ = pl.colorbar()
```

The matrix above shows the correlation values of movie states. The first three states are highly correlated while state 5 and state 4 are less similar to first three.

# State duration and probabilities for Movie condition

```python
from mvpa_itab.conn.states.utils import get_data, fit_centroids, get_transition_matrix, get_state_duration
from scipy.io import savemat
```

```python
state_duration_dict = dict()
transition_dict = dict()
```

```python
data, n_roi = get_data("/media/robbis/DATA/fmri/movie_viviana/masked/mat_corr_sub_MOVIE.mat")
fitted = fit_centroids(X=data, centroids=centers[0])
```

```python
transitions = get_transition_matrix(fitted)
```

```python
_, mean_subject_duration = get_state_duration(fitted)
```

```python
pl.figure(figsize=(10,8))
artist = pl.plot(mean_subject_duration[:,:,0].T, mean_subject_duration[:,:,1].T, marker='o', alpha=0.3)
artist2 = pl.plot(mean_subject_duration[:,:,0].mean(0), mean_subject_duration[:,:,1].mean(0), marker='o', lw=4.5, markersize=15)
aa=pl.xticks(range(1,6), range(1,6))
pl.xlabel("Movie state")
_ = pl.ylabel("Duration (frames)")
_ = pl.ylim(0, 100)
```

This Figure above indicates the average duration (in frames) of a state. Transparent rows represent each subject mean state duration.

```python
from matplotlib import colors
transitions[transitions==0] = 1
norm = colors.LogNorm()
pl.figure(figsize=(12,10))
pl.imshow(transitions/np.sum(transitions, axis=0), norm=norm)
pl.title("Transitions probability")
pl.xticks(np.arange(5),["Movie state %s" % str(i+1) for i in range(5)])
pl.yticks(np.arange(5),["Movie state %s" % str(i+1) for i in range(5)])
_ = pl.colorbar()
```

Plot above indicates the probability to change the state. The darker is the color the less probable is the transition between states in the rows and the states in the columns. 


```python
transition_p = transitions/np.sum(transitions, axis=0)
from scipy.linalg import eig
w, v = eig(transitions)
```

```python
pl.figure(figsize=(10,8))
pl.plot(np.abs(v[:,0]), marker='o', lw=2, markersize=19)
pl.xticks(range(0,5), range(1,6))
_ = pl.yticks(np.linspace(0,1,6), np.linspace(0,1,6))
_ = pl.xlabel("Movie state")
_ = pl.ylim(0, 1)
```

Plot above indicates the probability of states in the long run. In other words the stationar behaviour of the system.

```python
state_duration_dict['movie'] = mean_subject_duration
transition_dict['movie_probabilities'] = transition_p
transition_dict['movie_number'] = transitions
```
