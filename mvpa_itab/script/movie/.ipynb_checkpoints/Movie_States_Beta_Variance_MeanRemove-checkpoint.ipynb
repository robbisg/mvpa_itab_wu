# Metrics to choose the right number of states

## Movie condition state metrics

In the following plot we showed several metrics used in literature to choose the right number of states.
Usually in machine learning literature the number of state is choosen using the elbow method in the Silhouette or in the Global Explained Variance indices, The elbow method looks at the percentage of variance explained as a function of the number of clusters: One should choose a number of clusters so that adding another cluster doesn't give much better modeling of the data. 
Here comparing different index, it seems reasonable to choose 9 states, because in the Silhouette, KL and Explained Variance Graph adding a cluster to 9 does not outperform the indexes.

```python
from mvpa_itab.conn.states.plot import plot_center_matrix, plot_condition_centers, plot_metrics
```

```python
metrics_, k_step, metrics_keys = calculate_metrics(X['movie'], clustering['movie']);
```

```python
plot_metrics(metrics_=metrics_, k_step=k_step, metric_names=metrics_keys)
```

```python
from mvpa_itab.conn.states.utils import get_centroids
```

```python
node_number=[6,5,8,10,4,5,7]
node_networks = ['DAN','VAN','SMN','VIS','AUD','LAN','DMN']
position = [sum(node_number[:i+1]) for i in range(len(node_number))]
position_label = [-0.5+position[i]-node_number[i]/2. for i in range(len(node_number))]
```

```python
center_list = [get_centroids(X['movie'], c) for c in clustering['movie']]
```

```python
def get_states(center_list, k):
    sum_ = np.sum([i for i in np.arange(2, k)])
    return sum_, sum_+k
```

```python
choosen_k = 5
```

```python
b, e = get_states(center_list, k=choosen_k)
centroid_ts = TimeSeries(np.vstack(center_list)[b:e], sampling_interval=1)
S = SimilarityAnalyzer(centroid_ts, measure=correlation)
similarity = 1 - S.measure # Transformation from correlation distance to correlation coefficient
similarity_matrix = copy_matrix(similarity, diagonal_filler=1)
```

```python
plot_labels = ["Movie  %s" % str(i+1) for i in range(similarity_matrix.shape[0])]
```

```python
pl.figure(figsize=(10,8))
pl.imshow(similarity_matrix, cmap=pl.cm.RdBu_r, vmin=0)
pl.title("Correlation of Movie states")
pl.xticks(np.arange(similarity_matrix.shape[0]),plot_labels, rotation=45)
pl.yticks(np.arange(similarity_matrix.shape[0]),plot_labels)
_ = pl.colorbar()
```

```python
plot_condition_centers(X['movie'], clustering['movie'][choosen_k-2], vmax=None);
```
